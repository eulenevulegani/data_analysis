### DATA CLEANING IN PANDAS

```python
#import required library
import pandas as pd

# load data from excel file 
df = pd.read_excel(r"Customer Call List.xlsx")
df
```

```python
# drop duplicates
df = df.drop_duplicates()
df
```

```python
#drop columns that are not required
df = df.drop(columns= "Not_Useful_Column")
df
```

```python
#remove unnecessary characters fom Last_name entries
df["Last_Name"]= df["Last_Name"].str.strip("._/")
df
```

```python
#remove uneccesary characters from phone number entries
df["Phone_Number"] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]','')
df
```

```python
#change phone number format to standard format and remove null values
df["Phone_Number"] = df["Phone_Number"].apply (lambda x: str(x)[0:3] + '-' + str(x)[3:6] + '-'+ str(x)[6:10] )
df
```

```python
#remove null values from phone number column
df["Phone_Number"]= df["Phone_Number"].str.replace("nan--",'')
df["Phone_Number"]= df["Phone_Number"].str.replace("Na--",'')
df
```

```python
#split address into street, state and zip code
df[["Street_Address","State", "Zip_code"]]= df["Address"].str.split(",",2, expand = True)
df
```

```python
#make the boolean values uniform in paying customer column
df["Paying Customer"]= df["Paying Customer"]. str. replace ("Yes", "Y")
df["Paying Customer"]= df["Paying Customer"]. str. replace ("No", "N")
df
```

```python
#make the boolean values uniform in Do_not_contact column
df["Do_Not_Contact"]= df["Do_Not_Contact"]. str. replace ("Yes", "Y")
df["Do_Not_Contact"]= df["Do_Not_Contact"]. str. replace ("No", "N")
df
```

```python
#remove null values 
df= df.fillna('')
df
```

```python
for x in df.index:
    if df.loc[x, "Do_Not_Contact"] == 'Y':
        df.drop (x, inplace=True)
df
```

```python
#remove entries of customers to did notaccept 
for x in df.index:
    if df.loc[x, "Phone_Number"] == '':
        df.drop (x, inplace=True)
df
```

```python
#drop the split column
df= df.drop (columns = "Address")
df
```

```python
#reset index
df=df.reset_index(drop=True)
df
```
